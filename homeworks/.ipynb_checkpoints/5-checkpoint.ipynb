```python
n_groups = len(data['Educ'].unique())

f, ax = plt.subplots(nrows=2, ncols=n_groups, figsize=[15, 7])
ax_row = 0
for group, group_data in data.groupby('Educ'):
    ax[0][ax_row].set_title(group)
    sns.distplot(group_data['Income2005'], ax=ax[0][ax_row])
    sm.graphics.qqplot(group_data['Income2005'], dist=stats.t, line='r', fit=True, ax=ax[1][ax_row]);
    ax_row += 1
```
