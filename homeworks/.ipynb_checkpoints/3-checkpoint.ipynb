3. Two Sample T-Test:

---

1. Ho: ÂµF = ÂµNF | Ha: ÂµF â‰  ÂµNF
2. Draw & shade:

```python
df = len(fired) + len(non_fired) - 2
alpha = .05
t_critical = stats.t.ppf(1-alpha/2, df)
print('Critical Value t: Â±', t_critical)

x = np.linspace(stats.t.ppf(0.001, df), stats.t.ppf(0.999, df), 100)
y = stats.t.pdf(x, df)
plt.plot(x, y)
plt.fill_between(x=x, y1=0, y2=y * (np.abs(x) > t_critical).astype(int))
```

3.Test Statistic

```python
tstat, pvalue = stats.ttest_ind(fired, non_fired)
print('test statisic: ', tstat)
print('pvalue:', pvalue)
print('The probability that the two distributions have equal means is ', pvalue)
```

### A 95% confidence interval.

```python
from statsmodels.stats.weightstats import CompareMeans, DescrStatsW

cm = CompareMeans(DescrStatsW(fired), DescrStatsW(non_fired))
print(cm.tconfint_diff(alpha=0.05, alternative='two-sided', usevar='pooled'), 'years')
```

In the last homework I mentioned I polled my Business Stats class here at SMU and asked them how much money (cash) they had in their pocket at that very moment.  The idea was that we wanted to see if there was evidence that those in charge of the vending machines should include the expensive bill / coin acceptor or if it should just have the credit card reader.  However, I met a professor from Seattle University last year and asked her to poll her class with the same question.  Below are the results of our polls.

```python
smu = np.array([34, 1200, 23, 50, 60, 50, 0, 0, 30, 89, 0, 300, 400, 20, 10, 0])
seattle = np.array([20, 10, 5, 0, 30, 50, 0, 100, 110, 0, 40, 10, 3, 0])
```

Evidence Samples do not come from normal distributions:  There is significant evidence from the histograms and qqplots of severe departures from normality for these data.  We will assume they do not  come from normal distributions. In addition, the sample size does not look adequate to make the t-test robust to this assumption. 

Non - Equal standard deviation: There is significant evidence to suggest that the standard deviations of these distributions are different. 

Independence: We will assume that the observations are independent both between and within groups.

```python
f, ax = plt.subplots(nrows=1, ncols=2, figsize=[15, 7])
ax[0].set_title('smu')
sns.distplot(smu, ax=ax[0])
ax[1].set_title('seattle')
sns.distplot(seattle, ax=ax[1])
```

```python
f, ax = plt.subplots(nrows=1, ncols=2, figsize=[15, 7])
ax[0].set_title('smu')
sm.graphics.qqplot(smu, dist=stats.t, line='r', fit=True, ax=ax[0]);
ax[1].set_title('seattle')
sm.graphics.qqplot(seattle, dist=stats.t, line='r', fit=True, ax=ax[1]);
```

Since the assumptions of the t test are not met, we will instead conduct a permutation test for the difference in sample means.  

To test for a difference of population means between the SMU and Seattle groups, a permutation test was conducted on 1000 random permutations of the data.  A histogram of the 1000 differences of sample means from the 1000 permutation can be seen above.

```python
# obsorved difference
delta = abs(smu.mean() - seattle.mean())

pooled = np.hstack([smu, seattle])

n_permutations = 1000
extreme_count = 0
diffs = []
for _ in range(n_permutations):
    # just to get the same result
    np.random.seed(0)
    pooled = np.random.permutation(pooled)
    diff = pooled[:len(smu)].mean() - pooled[len(smu):].mean()
    diffs.append(diff)
    # count how many are more extreme then the obsorved difference
    if abs(diff) >= delta:
        extreme_count += 1

# calculate the p-value
pvalue = extreme_count / n_permutations
pvalue
```

```python
plt.title('Distribution of means')
plt.hist(diffs);
```

The observed difference was $114 in which 202 of the 1000 permutations yielded a difference in sample means that was as extreme or more extreme than this observed difference (pvalue = 202/1000 = .202).

This does not provide sufficient evidence against the null hypothesis that the mean pocket cash of SMU students is equal to that of Seattle University students.

These 30 students were not from a random sample therefore inference cannot be drawn beyond the 30 subjects in the sample.
